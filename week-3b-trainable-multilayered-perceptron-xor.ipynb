# Week 3b: Trainable Multilayered Perceptrons

## Import libraries

- `numpy` is used to manage arrays
- `keras` `Sequential` model and `Dense` layer are used for the model
- `matplotlib` is used for plotting

```python
from keras.models import Sequential
from keras.layers import Dense
import numpy as np 
import matplotlib.pyplot as plot
```

## Define variables

- `rows`, the dataset of the XOR function (false=0, true=1)
- `activation`, the activation function, here the step function
- `weights`, the weights of all three perceptrons

```python
x = np.array([[0,0],[0,1],[1,0],[1,1]])
y = np.array([[0],[1],[1],[0]])
```

## Prepare the model

```python
model = Sequential(
    [
        Dense(5, activation="sigmoid", name="hidden_layer"),
        Dense(1, activation="sigmoid", name="output_layer"),
    ]
)
model.compile(loss="mean_squared_error", optimizer="adam", metrics=["binary_accuracy"])
```

## Train the model

```python
model.fit(x, y, batch_size=4, epochs=1000)
```

## Plot the result

```python
plot.axis([-3, +4, -3, +4])

colors = { 0: 'blue', 1: 'red' }

# plot the rows
for row in zip(x, y):
    plot.plot(row[0][0], row[0][1], marker='o', color=colors[row[1][0]], markersize=5)

# plot the predictions
d = 40
px = np.array([[-3 + dx * 7 / d, -3 + dy * 7 / d] for dx in range(d) for dy in range(d)])
pp = model.predict([px])
for row in zip(px, pp):
    if row[1][0] < 0.5:
        plot.plot(row[0][0], row[0][1], marker='.', color=colors[0], markersize=1)
plot.tight_layout()
```
