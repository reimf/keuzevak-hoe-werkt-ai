# Week 3b: Trainable Multilayered Perceptrons

## Import libraries

- `numpy` is used to manage arrays
- `keras` `Sequential` model and `Dense` layer are used for the model
- `matplotlib` is used for plotting

```python
from keras.models import Sequential
from keras.layers.core import Dense
import numpy as np 
import matplotlib.pyplot as plot
```

## Define variables

- `rows`, the dataset of the XOR function (false=0, true=1)
- `activation`, the activation function, here the step function
- `weights`, the weights of all three perceptrons

```python
x = np.array([[0,0],[0,1],[1,0],[1,1]])
y = np.array([[0],[1],[1],[0]])
```

## Prepare the model

```python
model = Sequential()
model.add(Dense(2, activation='sigmoid', input_dim=2))
model.add(Dense(1, activation='sigmoid'))
model.compile(loss='mean_squared_error', optimizer='adam', metrics=['accuracy'])
```

## Train the model

```python
model.fit(x, y, batch_size=4, epochs=10000, verbose=0)
```

## Show predictions

```python
print(model.predict([x]))
```
